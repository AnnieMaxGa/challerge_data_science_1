# 2. Ventas por categoría

```python
import requests
import csv
import matplotlib.pyplot as plt


from collections import defaultdict
from io import StringIO

urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
    ]

ventas_por_tienda = {}
for url in urls:
    tienda_nombre = url.split("/")[-1].replace("%20", "_").split(".")[0]
    ventas_por_depto = defaultdict(int)

    response = requests.get(url)
    contenido_datos = response.text
    archivo_paso = StringIO(contenido_datos)
    lee_datos = csv.DictReader(archivo_paso)

    for fila in lee_datos:
      depto = fila['Categoría del Producto']
      ventas_por_depto[depto] += 1

    ventas_por_tienda[tienda_nombre] = dict(ventas_por_depto)

# RESULTADOS POR TIENDA
for tienda, ventas in ventas_por_tienda.items():
    print(f'\nVentas por categoría en {tienda}:')
    for categoria, ventas_categoria in ventas.items():
      print(f' {categoria}: {ventas_categoria}')
      #print()  # Línea en blanco entre tiendas

# MAS POPULAR
for tienda, ventas in ventas_por_tienda.items():
    categoria_popular = max(ventas.items(), key=lambda x: x[1])
    print(f"\nCategoría más Vendida en {tienda}: {categoria_popular[0]} ({categoria_popular[1]} ventas)")

# GRAFICO
tiendas = []
ventas_top = []

for tienda, ventas in ventas_por_tienda.items():
    categoria, cantidad = max(ventas.items(), key=lambda x: x[1])
    tiendas.append(tienda)
    ventas_top.append(cantidad)

# Gráfico de barras básico
plt.bar(tiendas, ventas_top)
plt.title('Categoría más vendida por tienda')
plt.xlabel('Tienda')
plt.ylabel('Cantidad de productos vendidos')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
import requests
import csv
import pandas as pd
import matplotlib.pyplot as plt

from collections import defaultdict
from io import StringIO

urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
    ]


calificacion_por_tienda = {}

for url in urls:
    tienda_nombre = url.split("/")[-1].replace("%20", "_").split(".")[0]
    califica = defaultdict(int)

    response = requests.get(url)
    contenido_datos = response.text
    archivo_paso = StringIO(contenido_datos)
    lee_datos = csv.DictReader(archivo_paso)

    for fila in lee_datos:
        calificacion = fila['Calificación']
        if calificacion:
            califica[calificacion] += 1

    calificacion_por_tienda[tienda_nombre] = dict(califica)

# RESULTADOS
for tienda, calificaciones in calificacion_por_tienda.items():
    print(f'\nCalificaciones en {tienda}:')
    for nota, cantidad in sorted(calificaciones.items()):
        print(f'  {nota}: {cantidad}')

# MAS ALTA Y MAS BAJA
for tienda, calificaciones in calificacion_por_tienda.items():
    calificacion_alta = max(calificaciones.items(), key=lambda x: x[1])
    calificacion_baja = min(calificaciones.items(), key=lambda x: x[1])
    print(f'\nCalificación más alta y más baja en {tienda}')
    print(f' La calificación más alta es: {calificacion_alta[0]}')
    print(f' La calificación más baja es: {calificacion_baja[0]}')

#GRAFICO
tiendas = []
calif_altas = []
calif_bajas = []

for tienda, calificaciones in calificacion_por_tienda.items():
    calificacion_alta = max(calificaciones.items(), key=lambda x: x[1])
    calificacion_baja = min(calificaciones.items(), key=lambda x: x[1])

    tiendas.append(tienda)
    calif_altas.append(calificacion_alta[1])
    calif_bajas.append(calificacion_baja[1])
# Crear gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(tiendas, calif_altas, marker='o', label='Calificación más alta', color='blue')
plt.plot(tiendas, calif_bajas, marker='o', label='Calificación más baja', color='green')

plt.title('Calificación más alta y más baja por tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
import requests
import csv
import pandas as pd
import matplotlib.pyplot as plt

from collections import defaultdict
from io import StringIO

urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
    ]


ventas_por_tienda = {}
producto_mas_vendido = {}
producto_menos_vendido = {}


for url in urls:
    tienda_nombre = url.split("/")[-1].replace("%20", "_").split(".")[0]
    ventas_por_producto = defaultdict(int)

    response = requests.get(url)
    contenido_datos = response.text
    archivo_paso = StringIO(contenido_datos)
    lee_datos = csv.DictReader(archivo_paso)

    for fila in lee_datos:
      productos = fila['Producto']

      ventas_por_producto[productos] +=1

    ventas_por_tienda[tienda_nombre] = dict(ventas_por_producto)

for tienda, productos in ventas_por_tienda.items():
    if not productos:
        print(f'Tienda: {tienda} no tiene datos de productos')
        continue

    producto_mas_vendido = max(productos.items(), key=lambda x: x[1])
    producto_menos_vendido = min(productos.items(), key=lambda x: x[1])

    print(f'\nTienda: {tienda}')
    print(f'Producto más vendido: {producto_mas_vendido[0]} Ventas: {producto_mas_vendido[1]}')
    print(f'Producto menos vendido: {producto_menos_vendido[0]} Ventas: {producto_menos_vendido[1]}')


# GRAFICO
tiendas = []
producto_mas_vendido = []
producto_menos_vendido = []
ventas_mas_vendido = []
ventas_menos_vendido = []

for tienda, productos in ventas_por_tienda.items():
    if not productos:
        continue

    #MAS Y MENOS VENDIDO
    producto_mas = max(productos.items(), key=lambda x: x[1])
    producto_menos = min(productos.items(), key=lambda x: x[1])

    tiendas.append(tienda)
    producto_mas_vendido.append(producto_mas[0])
    producto_menos_vendido.append(producto_menos[0])
    ventas_mas_vendido.append(producto_mas[1])
    ventas_menos_vendido.append(producto_menos[1])

fig, ax = plt.subplots(figsize=(8, 4))


bar_width = 0.30
index = range(len(tiendas))

bar1 = ax.bar(index, ventas_mas_vendido, bar_width, label='Más Vendido', color='blue')
bar2 = ax.bar([i + bar_width for i in index], ventas_menos_vendido, bar_width, label='Menos Vendido', color='green')


ax.set_xlabel('Tiendas')
ax.set_ylabel('Ventas')
ax.set_title('Producto Más y Menos Vendido por Tienda')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(tiendas, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()



```
